```python
# Data types

int_num = 123 # integer(int)
float_num = 123.456 # float
complex_num = 123 + 456j # complex
bool_var = True # (False) - boolean
str_var = "qwerty1234#@$#weEF#@" # string(str)

lst = [1, 2.456, -3, 'a', 'b', True] # list
tup = (1, 2.456, -3, 'a', 'b', True) # tuple
set_var = {1, 2.456, -3, 'a', 'b', True} # set
dict_var = {
    'name': 'John',
    'age': 20,
    'country': 'USA',
} # dictionary

rng = range(1, 10) # range(start, end, step) - from start to end with step
```

```python
# Operators

# Math
print(3 + 6) # add
print(67 - 21) # subtract
print(23 * 4) # mul
print(45 / 9) # truediv !return float!
print(2 ** 3) # power
print(45 // 5) # floordiv !return int!
print(10 % 3) # mod

# Compare (!return bool)
print(23 < 45) # less
print(12 > 33) # greater
print(45 <= 66) # less or equal
print(92 >= 23) # greater or equal
print(45 == 45) # equal
print(23 != 23) # not equal

# Logic

# and - если есть хотя бы одно False, результат будет False
print(True and True)   # True
print(True and False)  # False
print(False and True)  # False
print(False and False) # False

# or - если есть хотя бы одно True, результат будет True
print(True or True)    # True
print(True or False)   # True
print(False or True)   # True
print(False or False)  # False

# not - инверсия (переворачивает значение)
print(not True)        # False
print(not False)       # True


# Assignment (Присваивание)
a = 1
print(a)      # Выведет: 1

a += 2        # То же самое, что: a = a + 2
print(a)      # Выведет: 3

a -= 1        # То же самое, что: a = a - 1
print(a)      # Выведет: 2

a *= 3        # То же самое, что: a = a * 3
print(a)      # Выведет: 6

a /= 1        # То же самое, что: a = a / 1
print(a)      # Выведет: 6.0 (деление всегда возвращает float)
```

```python
# Conditions - условная конструкция (алгоритм ветвления)

num1 = int(input('Enter first number: '))
num2 = int(input('Enter second number: '))

if num1 > num2: # проверка первого условия
    print('First number')
elif num1 < num2: # если первое не подошло, проверяем это
    print('Second number')
else: # если ни одно из условий выше не сработало
    print('Numbers are equal')
```
